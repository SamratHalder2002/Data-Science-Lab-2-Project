```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
!pip install chardet
import chardet
with open('/content/drive/MyDrive/Users.csv', 'rb') as f:
    encoding = chardet.detect(f.read())['encoding']
dataset = pd.read_csv('/content/drive/MyDrive/Customer-Churn-Records.csv', encoding=encoding)
dataset.head()
```

```python
# Removing the 'Surname' and 'CustomerId' columns from the dataset
dataset = dataset.drop(columns=['Surname'])
dataset = dataset.drop(columns=['CustomerId'])

dataset
```

```python
# Creating dummy variables for the 'Geography' column
geography_dummies = pd.get_dummies(dataset['Geography'])

# Creating new columns 'IsFrance' and 'IsSpain'
dataset['IsFrance'] = geography_dummies['France']
dataset['IsSpain'] = geography_dummies['Spain']

# Assigning 0 to 'IsFrance' and 'IsSpain' where the country is neither France nor Spain
dataset['IsFrance'] = dataset['IsFrance'].fillna(0).astype(int)
dataset['IsSpain'] = dataset['IsSpain'].fillna(0).astype(int)

dataset
```

```python
# Removing the 'Geography	' column from the dataset
dataset = dataset.drop(columns=['Geography'])
dataset
```

```python
# Changing entries in the 'Gender' column
dataset['Gender'] = dataset['Gender'].replace({'Male': 1, 'Female': 0})
dataset
```

```python
# Checking entries in 'HasCrCard' and changing corresponding entries in 'Card Type'
dataset.loc[dataset['HasCrCard'] == 0, 'Card Type'] = 'NA'
dataset
```

```python
# Creating new columns 'IsGold', 'IsPlatinum', 'IsSilver', and 'IsDiamond'
dataset['IsGold'] = (dataset['Card Type'] == 'GOLD').astype(int)
dataset['IsPlatinum'] = (dataset['Card Type'] == 'PLATINUM').astype(int)
dataset['IsSilver'] = (dataset['Card Type'] == 'SILVER').astype(int)
dataset['IsDiamond'] = (dataset['Card Type'] == 'DIAMOND').astype(int)
dataset
```

```python
# Removing the 'Card Type' column from the dataset
dataset = dataset.drop(columns=['Card Type'])
dataset = dataset.drop(columns=['RowNumber'])
dataset
```

```python
# Importing necessary libraries for logistic regression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Selecting features and target variable
X = dataset[['CreditScore', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember',
             'EstimatedSalary', 'Complain', 'Satisfaction Score', 'Point Earned', 'IsFrance', 'IsSpain',
             'IsGold', 'IsPlatinum', 'IsSilver', 'IsDiamond']]
y = dataset['Exited']

# Splitting the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initializing and training the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Making predictions on the testing set
y_pred = model.predict(X_test)

# Calculating accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

accuracy
```
